# Robust Scaler
-If Your dataset contains more no of outliers then robut scalar is the best techique for scaling
### Robust Scalar formula
#      xl  -  Q2(x)
#      --------------
#      Q3(x)  - Q1(x)
### Inter Quatile Range(IQR)
#  Q1 --->25%(lowe Quarlite)
#  Q2---->50%(Medium Quarlite)
#   Q3---->75%(Upper Quarlite)

```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/Advertisement.csv')
```

```python
data1.head()
```

```python
data1.drop(['Unnamed: 0'],axis=1,inplace = True)
```

```python
data1.head()
```

# Apply Robust Model

```python
from sklearn.preprocessing import RobustScaler
```

```python
rs = RobustScaler()
rs
```

```python
rs2 = rs.fit_transform(data1)
rs2
```

```python
rs2 = pd.DataFrame(rs2, columns = data1.columns)
rs2.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2, figsize = (9,5))
sns.kdeplot(data1['TV'],ax = ax1)
sns.kdeplot(data1['radio'],ax = ax1)
sns.kdeplot(data1['newspaper'],ax = ax1)
sns.kdeplot(data1['sales'],ax = ax1)

sns.kdeplot(rs2['TV'],ax = ax2)
sns.kdeplot(rs2['radio'],ax = ax2)
sns.kdeplot(rs2['newspaper'],ax = ax2)
sns.kdeplot(rs2['sales'],ax = ax2)
```

# Standard Scaler
-Most commonly used scaling techique.It scales between 0 to mean

```python
### Standard Scaler Formula
#    xi - mean(x)
# -----------------
# std(x)
```

```python
from sklearn.preprocessing import StandardScaler
```
